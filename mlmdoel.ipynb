```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

messages = [
    "Congratulations you won a lottery",
    "Win cash now click here",
    "You have been selected for a free gift",
    "Hello how are you",
    "Are you coming to college",
    "Let's meet tomorrow",
    "Free entry in contest",
    "Call this number to claim your reward",
    "Can we meet today",
    "Please send the report","offer"
]

labels = [
    "spam",
    "spam",
    "spam",
    "not spam",
    "not spam",
    "not spam",
    "spam",
    "spam",
    "not spam",
    "not spam","spam"
]


cv = CountVectorizer()
X = cv.fit_transform(messages)


X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)


model = MultinomialNB()
model.fit(X_train, y_train)


user_input = input("Enter your message: ")


input_vec = cv.transform([user_input])


prediction = model.predict(input_vec)

print("\nPrediction:", prediction[0])
```
